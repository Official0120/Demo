**47 -** Write a Python program to combine values in python list of dictionaries. 
*  Sample data: [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount': 300},{'item': 'item1', 'amount': 750}]
*  Expected Output: Counter ({'item1': 1150, 'item2': 300})






```python
from collections import Counter
item_list = [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount': 300}, {'item': 'item1', 'amount': 750}]
result = Counter()
for d in item_list:
    result[d['item']] += d['amount']
print(result) 
```

**51 -** Write a Python function to check whether a number is perfect or not.

```python
def num(n):
    sum = 0
    for x in range(1, n):
        if n % x == 0:
            sum += x
    return sum == n
print(num(6))
```

**67 -** Write a Python program to find the maximum and minimum numbers from the specified decimal numbers.

```python
def max_min(num):
  l = num[0]
  s = num[0]
  for x in num:
    if x > l:
      l = x
    elif x < s:
        s = x
  return l, s 
  
print(max_min([0.1, 10.5, 4.15, 8.40, 9.0, 4.2, 17.0, 0.28, 0.75]))
```
